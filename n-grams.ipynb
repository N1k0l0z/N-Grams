```python
import requests
from bs4 import BeautifulSoup
def crawling(url):
    all_links = []
    def links_from_link(url):
        clean_links = []
        response = requests.get(url)
        soap = BeautifulSoup(response.content)
        main = soap.find_all('p')
        for paragraph in main:
            for link in paragraph.find_all('a'):
                href = link.get('href')
                if href:
                    if not href.startswith('http'):
                        clean_links.append('https://en.wikipedia.org' + href)
                    else:
                        clean_links.append(href)
        return clean_links
    def link_depth(url, depth):
        all_links.append(url)
        for l in range(depth):
            current_links = list(all_links)
            for link in current_links:
                new_links = links_from_link(link)
                all_links.extend(new_links)
    depth = 1
    link_depth(url, depth)
    def read_text(url):
        response = requests.get(url)
        soap = BeautifulSoup(response.content)
        text = soap.find_all('p')
        # Handle cases where 'h1' tag might be missing
        title_elements = soap.find_all('h1')
        title = title_elements[0].text if title_elements else "Title Not Found"
        return title, text
    data = []
    for i in all_links:
        title, text_list = read_text(i)
        for k in range(len(text_list)):
            data.append([text_list[k].text])
    return data
```

```python
def tech_page_scrapping(url):
  response = requests.get(url)
  soap = BeautifulSoup(response.content)
  main = soap.find_all('p')
  clean_links = []
  for paragraph in main:
    for link in paragraph.find_all('a'):
        href = link.get('href')
        if href:
            if not href.startswith('http'):
                clean_links.append('https://en.wikipedia.org' + href)
            else:
                clean_links.append(href)
  return clean_links[33:53] # 416
tech_links = tech_page_scrapping('https://en.wikipedia.org/wiki/Wikipedia:Good_articles/Engineering_and_technology#Computing_and_engineering/')
finished_data = []
for i in tech_links:
  finished_data.append(crawling(i))
```

```python
import re

def clean_wikipedia_text(data):
    cleaned_data = []
    for item in data:

        flattened_item = [text for sublist in item for text in sublist]
        for text in flattened_item:
            text = re.sub(r'\[.*?\]', '', text)

            text = re.sub(r'\(.*?\)', '', text)

            text = text.replace('\n', ' ')

            text = re.sub(r'[^a-zA-Z0-9\s]', '', text)

            text = re.sub(r'\s+', ' ', text).strip()

            if text:
                cleaned_data.append(text)

    return cleaned_data
```

input
*   n: The order of the n-gram model (in this case, 3 for trigrams).
*   tokenized_text: A list of tokenized sentences, e.g., [['this', 'is', 'a',
    'sentence'], ['this', 'is', 'another']].
padded_sents: List of padded sentences.
[
  ['<s>', '<s>', 'this', 'is', 'a', 'sentence', '</s>'], ['<s>', '<s>', 'this', 'is', 'another', '</s>']
]
train_data: List of trigrams
[
  [('<s>', '<s>', 'this'), ('<s>', 'this', 'is'), ('this', 'is', 'a'), ('is', 'a', 'sentence'), ('a', 'sentence', '</s>')],
  [('<s>', '<s>', 'this'), ('<s>', 'this', 'is'), ('this', 'is', 'another'), ('is', 'another', '</s>')]
]


```python
from nltk.lm.preprocessing import padded_everygram_pipeline
n = 3
train_data, padded_sents = padded_everygram_pipeline(n, tokenized_text)
```

without smoothing:
P("was" | "this") = 0

With Laplace smoothing:
P("was" | "this") = (0 + 1) / (total bigrams from "this" + vocabulary size)

```python
from nltk.lm import Laplace
model = Laplace(1)
```

```python
model.fit(train_data, padded_sents)
```

```python
print(model.generate(10, text_seed=['Bill']))
```

```python
import random
from nltk.tokenize import word_tokenize
from nltk.lm import Laplace
import nltk

nltk.download('punkt')

def generate_sent(model, num_words, random_seed=None, text_seed=None):
    content = []
    if text_seed is not None:

        text_seed_string = ' '.join(text_seed)
        content.extend(word_tokenize(text_seed_string))

    if random_seed is not None:
        random.seed(random_seed)

    generated_tokens = model.generate(num_words, text_seed=content)

    cleaned_tokens = [token.lower() for token in generated_tokens if token.isalnum()]

    content.extend(cleaned_tokens)
    return " " .join(content)
```
